```python
import pandas as pd
import numpy as np
import ast
```

```python
df_reviews = pd.read_csv('./data/reviews.csv', encoding = 'utf-8')
df_reviews.head()
```

```python
df = df_reviews.copy()
```

```python
# Filter for only string values that contain the dictionary string
mask = df['style'].apply(lambda x: isinstance(x, str) and "{'Format:" in x)
df = df.loc[mask]

# Convert the string dictionary to a dictionary data type
df['style'] = df['style'].apply(lambda x: ast.literal_eval(x))

# Extract the value for the dictionary key 'Format:' for each row
df['style'] = df['style'].apply(lambda x: x['Format:'])

# Assign this extracted value back to the 'style' column in the dataframe
df['style'] = df['style'].astype(str)

df.head()
```

```python
dfnew = df[df['style'].str.contains('Amazon Video|Prime Video')]
dfnew.info()
```

```python
df_features = dfnew[dfnew['verified'] == True]
df_features['asin'].value_counts()
```

```python
df_features = df_features.drop(['image', 'verified', 'unixReviewTime'], axis=1)
```
