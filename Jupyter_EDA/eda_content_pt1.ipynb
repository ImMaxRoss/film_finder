```python
df['category'] = df['category'].apply(lambda x: ast.literal_eval(x))
df['category'] = df['category'].apply(lambda x: tuple(x) if not isinstance(x, str) else x)

df['also_buy'] = df['also_buy'].apply(lambda x: ast.literal_eval(x))
df['also_buy'] = df['also_buy'].apply(lambda x: tuple(x) if not isinstance(x, str) else x)

df['also_view'] = df['also_view'].apply(lambda x: ast.literal_eval(x))
df['also_view'] = df['also_view'].apply(lambda x: tuple(x) if not isinstance(x, str) else x)

df.drop_duplicates(inplace=True)
df.head()
```

```python
df.info()
```

```python
df['category'] = df['category'].apply(lambda x: list(x) if isinstance(x, tuple) else x)

df['also_buy'] = df['also_buy'].apply(lambda x: list(x) if isinstance(x, tuple) else x)

df['also_view'] = df['also_view'].apply(lambda x: list(x) if isinstance(x, tuple) else x)

df['details'] = df['details'].replace(np.NaN, '[{}]')
df['details'] = df['details'].apply(lambda x: ast.literal_eval(x.strip('[]')))

df['description'] = df['description'].apply(lambda x: " ".join(ast.literal_eval(x)).strip())
```

```python
df.head()
```

```python
df.to_csv('./data/meta_content.csv', encoding='utf-8', index=False)
```
